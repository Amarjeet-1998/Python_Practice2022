Write a function dict_merge_sum that takes two dictionaries d1 and d2 as parameters. The values of both dictionaries are numerical. The function should return the merged sum dictionary m of those dictionaries. If a key k is both in d1 and d2, the corresponding values will be added and included in the dictionary m If k is only contained in one of the dictionaries, the k and the corresponding value will be included in m

```python
def dict_merge_sum(d1,d2):
  ''' Merging and calculating the sum of two dictionaries:
  Two dictionaries d1 and d2 with numerical values and possibly disjoint keys are merged and the values are added if 
  the exist in both values, otherwise the missing value is taken to be 0'''

  merged_sum = d1.copy()
  for key, value in d2.items():
    if key in d1:
      d1[key] += value
    else:
      d1[key] = value
  return merged_sum
```

In the Gregorian calendar, each leap year has 366 days instead of 365, by extending February to 29 days rather than the common 28. These extra days occur in years which are multiples of four (with the exception of centennial years not divisible by 400). Write a Python program, which asks for a year and calculates, if this year is a leap year or not.

```python



# Estimating the slope (B1)

def mean(p):
    """This function calculates the average value of given list
    :param p: a list of floating point values
    :return: a floating point value
    """
    return sum(p) / len(p)


def residual(q, mean_value):
    """This function calculates residual of each value of given list from mean
    :param q: a list of floating point values
    :param mean_value: float
    :return: a list of floating point values
    """
    residuals_list = []
    for i in q:
        residuals_list.append(i - mean_value)
    return residuals_list


def multiply(list_1, list_2):
    """This function multiplied each corresponding value in given lists
    :param list_1: a list of floating point values
    :param list_2: a list of floating point values
    :return: a list of floating point values
    """
    multiplied = []
    for i, j in zip(list_1, list_2):
        multiplied.append(i * j)
    return multiplied


def square(s):
    """This function squares all elements of given list
    :param s: a list of floating point values
    :return: a list of floating point values
    """
    squared = []
    for i in s:
        squared.append(i ** 2)
    return squared


def slope(ml, sq):
    """This function calculates the value of slope
    :param ml: a list of floating point values
    :param sq: a list of floating point values
    :return: a floating point value
    """
    return sum(ml) / sum(sq)


# Estimating the intercept (B0)

def intercept(mean_x_val, mean_y_val,  slope_val):
    """This function calculates the value of intercept
    :param mean_x_val: a floating point value
    :param mean_y_val: a floating point value
    :param slope_val: a floating point value
    :return: a floating point value
    """
    return mean_y_val - (slope_val * mean_x_val)


# Make Predictions

def predict(slope_val, intercept_val, x):
    """This function predicts the 'y' value according to the 'x'
    :param slope_val: a floating point value
    :param intercept_val: a floating point value
    :param x: a floating point value or a list of floating point values
    :return: a floating point value or a list of floating point values
    """
    if type(x) != list:
        pre_y = intercept_val + (slope_val * x)
        return pre_y

    elif type(x) == list:
        pre_y = []
        for i in x:
            pre_y.append(intercept_val + (slope_val * i))
        return pre_y

    else:
        raise TypeError("Type of argument should be int, float , list")


# Estimating Error

def RMSE(pre_y, actual_y, pre_x, actual_x):
    """This function calculates the error across all predicted values
    :param pre_x: a floating point value
    :param actual_x: a list of floating point values
    :param pre_y: a floating point value or a list of floating point values
    :param actual_y: a floating point value or a list of floating point values
    :return: a floating point value
    """
    if type(pre_y) != list:
        if pre_x in actual_x:
            index = actual_x.index(pre_x)
            subtracted = pre_y - actual_y[index]
            error = subtracted ** 2
            return error
        else:
            raise AssertionError("Your entered x to predict y is not in list")

    elif type(pre_y) == list:
        subtracted = []
        for i, j in zip(pre_y, actual_y):
            subtracted.append(i - j)

        for i in range(len(subtracted)):
            subtracted[i] = subtracted[i] ** 2

        item_sum = 0
        for i in subtracted:
            item_sum += i

        return item_sum / len(subtracted)

    else:
        raise TypeError


def main():
    """This function starts execution of program"""

    print(" Simple Linear Regression ".center(40, "*"))

    # this causes that two input list has same length
    data_length = int(input("Please Enter The Length of your data : "))

    x_in = []  # x indices of our input data
    y_in = []  # y indices of our input data

    print("\n", "*" * 39, "\n")

    for i in range(data_length):
        x_in.append(float(input("Please Enter Your x Data : ")))

    print("\n", "*" * 39, "\n")

    for j in range(data_length):
        y_in.append(float(input("Please Enter Your y Data : ")))

    mean_x = mean(x_in)  # mean value of 'x_in' list
    mean_y = mean(y_in)  # mean value of 'y_in' list

    # Residual of each x value from the mean_x
    res_x = residual(x_in, mean_x)
    # Residual of each y value from the mean_y
    res_y = residual(y_in, mean_y)

    # Multiplication of the x and y residuals from their means
    multiplied = multiply(res_x, res_y)

    # Squared residual of each x value from the mean
    squared = square(res_x)

    # calculate the value of slope
    B1 = slope(multiplied, squared)

    # calculate the value of intercept
    B0 = intercept(mean_x, mean_y, B1)

    print("\nThe Value of Slope is : ", B1)
    print("The Value of Intercept is : ", B0, "\n")

    print(" Make Predictions ".center(40, "*"))
    print("*" * 40)

    data_length = int(input("Please Enter The Length of your data : "))
    print("\n", "*" * 39, "\n")

    if data_length == 1:
        pre_x = float(input("Please Enter Your x data to predict : "))
        predicted_y = predict(B1, B0, pre_x)
        print("\nPredicted Y :", predicted_y, "\n")

    elif 1 < data_length <= len(x_in):
        pre_x = []
        for i in range(data_length):
            pre_x.append(float(input("Please Enter Your x data to predict : ")))
        predicted_y = predict(B1, B0, pre_x)
        print("\nPredicted Ys :", predicted_y, "\n")

    elif data_length > len(x_in):
        raise ValueError("Data length should not be greater than length of primitive list(x_in or y_in)")

    else:
        raise ValueError("Data length should be positive number")

    print("*" * 40)

    print(" Estimating Error ".center(40, "*"))
    print("\nSquared error :", RMSE(predicted_y, y_in, pre_x, x_in))
    print("Process Finished!")


if __name__ == '__main__':
    main()
```
